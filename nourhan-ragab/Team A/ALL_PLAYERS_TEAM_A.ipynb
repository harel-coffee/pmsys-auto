```python
%cd ../utils
from data_utils import *
from framework_utils import *
from tsai.all import *
import pandas as pd
import time as t
%run framework_utils.py
%run data_utils.py
```

```python
df_2021 = read_file_to_dataframe('../data/team_a_2021.xlsx')
df_2020 = read_file_to_dataframe('../data/team_a_2020.xlsx')

player_id = '25d82d44-93bb-41c6-8557-fccf35bf659f'

df_2020 = pd.DataFrame.from_dict(df_2020['Readiness'].astype(float))
df_2021 = pd.DataFrame.from_dict(df_2021['Readiness'].astype(float))

start = 50
end = 200

df_player_2020_test = df_2020[player_id].iloc[start:end]
df_player_2021_test = df_2021[player_id].iloc[start:end]


df_2020.drop(df_2020.index[start:end], inplace=True)
df_2021.drop(df_2021.index[start:end], inplace=True)

df_player_2020_train = df_2020[player_id]
df_player_2021_train = df_2021[player_id]

df_2020 = df_2020.drop([player_id], axis=1)
df_2021 = df_2021.drop([player_id], axis=1)

df_one_year_2020 = pd.concat([df_2020, df_player_2020_train], axis=1)
df_one_year_2021 = pd.concat([df_2021, df_player_2021_train], axis=1)


df_one_year_2020_concat = df_one_year_2020.melt().drop('variable',axis=1).dropna()
df_one_year_2020_concat.index = df_one_year_2020_concat.index + 1


df_one_year_2021_concat = df_one_year_2021.melt().drop('variable',axis=1).dropna()
df_one_year_2021_concat.index = df_one_year_2021_concat.index + 1


df_two_years_team_a = pd.concat([df_one_year_2020, df_one_year_2021])
df_two_years_team_a_concat = df_two_years_team_a.melt().drop('variable',axis=1).dropna()
df_two_years_team_a_concat.index = df_two_years_team_a_concat.index + 1

df_two_years_test = pd.concat([df_player_2020_test])

epoch = 1
batch_size = 5
n_splits = 1
shuffle = False
predictionWindowSize = 7
```
